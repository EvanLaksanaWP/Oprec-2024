## DataFrame

DataFrame pada Pandas adalah sebuah struktur data dua dimensi yang mirip dengan tabel terdiri dari baris dan kolom.

Komponen pada DataFrame:

1. Data, Nilai-nilai yang disimpan di dalam DataFrame dengan tipe data yang dapat beragam
2. Index, Label unik pada setiap baris pada DataFrame
3. Columns Names, Label atau identifikasi terhadap data yang disimpan dalam suatu kolom
4. Data Types, Tipe data dari setiap kolom pada DataFrame


```python
data
```

### Membuat DataFrame Secara Manual

Untuk memahami lebih lanjut, DataFrame juga dapat dibuat dengan menggunakan bantuan struktur data **Dictionary** pada Python.

- **Key** interpretasi dari label kolom dengan tipe data string
- **Value** interpretasi dari data pada suatu kolom dengan tipe data list
- Label index secara otomatis dibuat dari 0 hingga n


```python
dict_data = {
    'Nama': ['Evan', 'Firhan', 'Byu', 'Mira'],
    'Jenis Kelamin': ['L', 'L', 'L', 'P'],
    'Tempat Tinggal': ['Banyuwangi', 'Bandung', 'Malang', 'Malang'],
    'Umur': [20, 21, 20, 21]
}

manual_df = pd.DataFrame(dict_data)
manual_df
```

### Modifikasi Komponen pada DataFrame


```python
# Mengubah label kolom pada DataFrame
manual_df = manual_df.rename(columns={
                             'Nama': 'Name', 'Jenis Kelamin': 'Gender', 'Tempat Tinggal': 'Address', 'Umur': 'Age'})

# Mengubah label indeks pada DataFrame
manual_df.rename(index={2: 'index baru'}, inplace=True)

manual_df
```

## Series

Series dalam Pandas adalah struktur data satu dimensi yang mirip dengan array atau kolom dalam tabel. Series dapat menyimpan data dari berbagai tipe, seperti integer, float, string, atau objek Python. Satu hal yang membedakan Series dari array sederhana adalah adanya indeks yang menyertai setiap elemen dalam Series.

Komponen pada Series:

1. Data, Nilai-nilai yang disimpan dalam Series yang dapat berupa berbagai jenis tipe data.
2. Index, Label unik pada setiap elemen pada Series, secara default akan diberikan index label dari 0 hingga n.


```python
# Mengambil Series dari kolom Age pada Dataframe data
age_series = data["Age"]

print("Tipe data:", type(age_series))

age_series
```

### Seleksi

Mendapatkan nilai pada baris atau kolom tertentu pada DataFrame


#### Seleksi menggunakan [ ]

Indexing paling sederhana untuk mengambil data pada suatu kolom namun tidak bisa mengambil pada baris tertentu.

**data[columns_name]**


```python
# Mengambil semua data pada sebuah kolom tertentu dalam bentuk DataFrame
data[["Survived"]]
```

```python
# Mengambil semua data pada sebuah kolom tertentu dalam bentuk Series
data["Survived"]
```

```python
# Mengambil semua data pada kolom tertentu
data[["Name", "Sex", "Age"]]
```

#### Seleksi Menggunakan loc

Indexing menggunakan **label/nama** baris dan kolom

**data.loc[row_label, column_label]**


```python
# Mengambil baris index 0 hingga 5 pada semua kolom
data.loc[0:5, ]
```

```python
# Mengambil semua baris pada kolom tertentu
data.loc[:, ["Name", "Cabin"]]
```

```python
# Mengambil range index baris ternetu pada kolom tertentu menggunakan list
data.loc[887:891, ["Name", "Age"]]
```

```python
# Mengambil range index baris ternetu pada range kolom tertentu
data.loc[0:5, "Name":"Age"]
```

```python
# Mengambil baris index tertentu dan pada kolom tertentu menggunakan list
data.loc[[1, 5, 881, 886], ['Name', 'Age']]
```

#### Seleksi Menggunakan iloc

Indexing menggunakan **index integer**, cara penggunaan mirip dengan loc namun menggunakan index integer

**data.iloc[row_index, column_index]**


```python
# Mengambil range baris tertentu pada range kolom tertentu
data.iloc[1:5, 0:5]
```

```python
# Mengambil semua baris pada kolom tertentu
data.iloc[:, [0, 5]]
```

#### Seleksi Index Boolean

.loc dan [ ] dapat digunakan sebagai **Boolean Indexing**


```python
# Mengembalikan hasil operasi boolean dalam bentuk struktur data 1 dimensi
filter_umur_balita = data["Age"] < 5
filter_umur_balita
```

```python
# Mengembalikan baris data yang bernilai True atau memenuhi kondisi kolom Age < 5 (Balita)
data[filter_umur_balita]
```

```python
# Menampilkan baris data dengan Age < 5 dan Survived == 0

data[(data["Age"] < 5) & (data["Survived"] == 0)]
```

```python
# Menampilkan baris data dengan Age < 5 atau Survived == 0

data[(data["Age"] < 5) | (data["Survived"] == 0)]
```

## Pandas untuk Data Cleaning


### Nilai Kosong

**isna()** Akan mengembalikan **True** jika sel bernilai **null/Nan**


```python
data.isna()
```

```python
# sum() akan menampilkan banyak null tiap kolom
null_per_column = data.isna().sum()
null_per_column
```

### Membersihkan Nilai Kosong


```python
data.dropna(inplace=True)
data.isna().sum()
```

### Data Duplikat

**duplicated()** mengembalikan **True** jika baris terduplikasi


```python
data.duplicated().sum()
```

### Membersihkan Data Duplikat


```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

### Menghapus Baris & Kolom


```python
copy_data = data.copy()
copy_data.head()
```

**labels**

- _single label or list-like_: Index or column labels to drop. A tuple will be used as a single label and not treated as a list-like.

**axis**

- _{0 or 'index', 1 or 'columns'}_, default 0: Whether to drop labels from the index (0 or 'index') or columns (1 or 'columns').

**index**

- _single label or list-like_: Alternative to specifying axis (labels, axis=0 is equivalent to index=labels).

**columns**

- _single label or list-like_: Alternative to specifying axis (labels, axis=1 is equivalent to columns=labels).


```python
copy_data.drop(index=[2, 4], axis=1, inplace=False)
```
